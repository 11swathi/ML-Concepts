# How to plot a sinusoid function

The following Python code plots the function $f(x)=sin(x)$ for the interval $x \in [0,3]$. Click on the triangle (top, left) of the following cell to run this code:

```python
import numpy as np                     # this command imports the numpy (numerical Python) library, and gives it the nickname "np"
import matplotlib.pyplot as plt        # this command imports the matplotlib plotting library, and gives it the nickname "plt"
x = np.array([0,.5,1.,1.5,2.,2.5,3.])  # this command uses the numpy library to create a vector of 7 elements from 0 to 3
y = np.sin(x)                          # this command uses the numpy library to create a vector of 7 elements that contains sin(x) for each element of vector x
plt.figure()                                 # this command uses the plotting library to open a box for plotting
plt.plot(x,y, label='sin(x)')                # this command uses the plotting library to instruct the horizontal axis to be the x vector, and the vertical axis to be the y vector
plt.xlabel('x')                              # this command uses the plotting library to insert a label "x" for the horizontal axis
plt.ylabel('sin(x)')                         # this command uses the plotting library to insert a label "sin(x)" for the vertical axis
plt.title('not a very good plot of sine')    # this command uses the plotting library to insert a title for the plot
plt.grid()                                   # this command uses the plotting library to insert a grid for the plot
plt.legend()                                 # this command uses the plotting library to insert a legend for the plot (useful when you plot two or more curves at once)
```

This plot looks a bit choppy. The plot would be smoother if we used more samples, than just 7. Next, we try to do a slightly better job plotting both a  $sin(x)$ and a $cos(x)$. This time, we'll use 1000 points instead of 7.

Please note that we have already imported the two libraries numpy and matplotlib so we do not need to re-import them. If you are writing a brand new program you need to import these two libraries again.

```python
import numpy as np                     # this command imports the numpy (numerical Python) library, and gives it the nickname "np"
import matplotlib.pyplot as plt
no_samples = 1000                               # Here we define the number of samples for our plot
x = np.linspace(0,2*np.pi, no_samples)          # this command uses the numpy library to create a vector with smallest number 0 and biggest number 2 pi, and the 1000 values are evenly distributed between those two values
y_sin = np.sin(x)                               # this command uses the numpy library to create a vector of 1000 elements that contains sin(x) for each element of vector x
y_cos = np.cos(x)                               # this command uses the numpy library to create a vector of 1000 elements that contains cos(x) for each element of vector x
plt.figure()                                          # this command uses the plotting library to open a box for plotting
plt.plot(x,y_sin, label='sin(x)')                     # this command uses the plotting library to instruct the horizontal axis to be the x vector, and the vertical axis to be the y_sin vector
plt.plot(x,y_cos, label='cos(x)')                     # this command uses the plotting library to instruct the horizontal axis to be the x vector, and the vertical axis to be the y_cos vector
plt.xlabel('x')
plt.ylabel('sin and cos(x)')
plt.title(' a better plot of sine and cosine')
plt.grid()
plt.legend()                                          # this command uses the plotting library to insert a legend for the plot. Note that the labels from the plt.plot() above helps distinguish the two curves
```

So far, we have plotted two clean curves. In practical situations, our observations of data will be corrupted (perturbed) by noise. Very often, the noise will be additive and will have a Gaussian distribution. Python allows us to have an idea of how this looks by creating a vector of realizations of a random variable with whatever probability distribution we want. Then we can add these "noise" vectors to our clean plots, so each element of a vector is perturbed.

Here, we plot the same $sin(x)$ and $cos(x)$, but we add some random numbers to each point:

```python
rand_1 = np.random.normal(0,0.2, no_samples)  # this command uses the numpy library to create a vector of 1000 elements, each with a number from a Gaussian distribution (mean 0 and standard deviation 0.2)
rand_2 = np.random.normal(0,0.5, no_samples)  # this command uses the numpy library to create another vector of 1000 elements, each with a number from a Gaussian distribution (mean 0 and standard deviation 0.5)
y_sin_plus_noise = y_sin+rand_1             # this command to add the first vector of noise realization to the sin(x) vector
y_cos_plus_noise = y_cos+rand_2             # this command to add the second vector of noise realization to the cos(x) vector
plt.figure()
plt.plot(x,y_sin_plus_noise, label='sin(x) w/ noise standard deviation 0.2')
plt.plot(x,y_cos_plus_noise, label='cos(x) w/ noise standard deviation 0.5')
plt.xlabel('x')
plt.ylabel('noisy sin and noisier cos(x)')
plt.title(' noisy sine and cosine')
plt.grid()
plt.legend()
```

When programming, it is always best to set variables to number values, as opposed to hard-coding them into the functions. In this case, we'd like to set the mean and standard deviation to variables, to make them easy to change without having to hunt them down through the code.

Let us make the data a little less noisy:

```python
mean = 0                                            # we'll use this variable to hold the Gaussian mean
sigma = .1                                          # we'll use this variable to hold the Gaussian standard deviation (much smaller then before)
rand_1 = np.random.normal(mean,sigma, no_samples)   # this command uses the numpy library to create a vector of 1000 elements, each with a number from a Gaussian distribution (mean 0 and standard deviation 0.1)
rand_2 = np.random.normal(mean,sigma, no_samples)
y_sin_plus_noise = y_sin+rand_1
y_cos_plus_noise = y_cos+rand_2
plt.figure()
plt.plot(x,y_sin_plus_noise, label='sin(x) w/ noise standard deviation 0.1')
plt.plot(x,y_cos_plus_noise, label='cos(x) w/ noise standard deviation 0.1')
plt.xlabel('x')
plt.ylabel('sin and cos(x)')
plt.title(' noisy sine and cosine')
plt.grid()
plt.legend()
```

```python
#(Problem 1) Plot exactly 10 periods of sin(3x) and cos(3x) (Note that x is scaled here).

#Answer:

import numpy as np
import matplotlib.pyplot as plt

no_samples = 1000                               # Number of samples for our plot
x = np.linspace(0,20*np.pi, no_samples)         # 10 periods * 2*pi, so 0 to 20*pi
#x = np.linspace(-3*np.pi,3*np.pi, no_samples)
y_sin = np.sin(3*x)
y_cos = np.cos(3*x)

plt.figure(figsize=(20, 5))                      # Set the figure size

# Plot sin(3x)
plt.subplot(2, 1, 2)                            # Plot in the Sine wave
plt.plot(x, y_sin, label='sin(3x)')

# Plot cos(3x)
plt.subplot(2, 1, 2)                            # Plot in the Cos wave
plt.plot(x, y_cos, label='cos(3x)')

plt.xlabel('x')                                 #Label the axis
plt.ylabel('sin3(x) and cos3(x)')
plt.title('10 periods of sin3(x) and cos3(x)')

plt.grid(True)
plt.legend()
plt.tight_layout()                              # Adjust layout
plt.show()                                      # Show the plot
```
